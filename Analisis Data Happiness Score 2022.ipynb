```python
import pandas as pd
import numpy as np

import os

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
folder_path = "/content/drive/MyDrive/Data CSV"
```

```python
os.listdir(folder_path)
```

```python
path_file_2022 = os.path.join(folder_path, '2022.csv')

df_2022 = pd.read_csv(path_file_2022)
```

```python
df_2022.tail()
```

```python
df_2022.info()
```

```python
df_2022 = df_2022.dropna()
```

```python
df_2022.columns.tolist()
```

```python
obj_col = df_2022.select_dtypes(include=['O']).columns

columns = [col for col in obj_col if col not in ['RANK', 'Country']]

for col in columns:
    df_2022[col] = df_2022[col].str.replace(',', '.').astype(float)
```

```python
colors_green = ["#004D00", "#197419", '#4CAF50', '#80C080', '#C0E0C0']
colors_dark = ["#1F1F1F", "#313131", '#636363', '#AEAEAE', '#DADADA']
colors_red = ["#331313", "#582626", '#9E1717', '#D35151', '#E9B4B4']
colors_mix = ["#17869E", '#264D58', '#179E66', '#D35151', '#E9DAB4', '#E9B4B4', '#D3B651', '#6351D3']
colors_div = ["#004D00", '#4CAF50', '#DADADA', '#D35151', '#331313']

sns.palplot(colors_green)
sns.palplot(colors_dark)
sns.palplot(colors_red)
sns.palplot(colors_mix)
sns.palplot(colors_div)
```

```python
Negara_Maju =  ["United States", "United Kingdom", "Canada", "Germany", "Australia", "France", "Switzerland", "Netherlands", "Sweden", "Norway", "Denmark","Japan","Singapore","South Korea","Austria","Ireland","Belgium","Finland","New Zealand"]
```

```python
df_2022.loc[df_2022["Country"].isin(Negara_Maju)]
```

```python
df = df_2022[df_2022['Country'].isin(Negara_Maju)]

fig, ax = plt.subplots(figsize=(14, 8))

bars1 = ax.bar(df['Country'], df['Happiness score'], color=colors_dark[3], alpha=0.4, edgecolor=colors_dark[0])

ax.set_xticklabels(df['Country'], rotation=45, ha="right")
ax.set_xlabel("Countries", fontsize=10, labelpad=10, fontweight='bold', color=colors_dark[0])
ax.set_ylabel("Happiness Score", fontsize=14, labelpad=5, fontweight='bold', color=colors_dark[0])

plt.show()
```

```python
df = df_2022[df_2022['Country'].isin(Negara_Maju)]
median_score = df_2022['Happiness score'].median()
fig, ax = plt.subplots(figsize=(14, 8))
bars1 = ax.bar(df['Country'], df['Happiness score'], color=colors_green[3], alpha=0.4, edgecolor=colors_dark[0])
line  = ax.axhline(median_score, linestyle='--', color=colors_dark[2])
ax.set_xticklabels(df['Country'], rotation=45, ha="right")
ax.legend(["Median", "Past Average"], loc='upper center', bbox_to_anchor=(0.5, -0.3), ncol=4, borderpad=1, frameon=False, fontsize=12)
ax.grid(axis='y', alpha=0.3)
ax.set_axisbelow(True)
ax.set_xlabel("Countries", fontsize=14, labelpad=10, fontweight='bold', color=colors_dark[0])
ax.set_ylabel("Happiness Score", fontsize=14, labelpad=10, fontweight='bold', color=colors_dark[0])
xmin, xmax = ax.get_xlim()
ymin, ymax = ax.get_ylim()
plt.text(s="Dampak Kemajuan Negara terhadap Kebahagiaan Warga Negara", ha='left', x=xmin, y=ymax*1.12, fontsize=24, fontweight='bold', color=colors_dark[0])
plt.title(f"Dalam konteks negara maju, kemajuan negara memengaruhi kesejahteraan dan kebahagiaan rakyat secara langsung dan signifikan.\nKarena semua menempati posisi diatas garis median.", loc='left', fontsize=12, color=colors_dark[2])
plt.tight_layout()
plt.show()
```

```python
countries = ['Finland', 'South Korea']
df_plot = df_2022[df_2022['Country'].isin(countries)]
for i , r in df_plot.iterrows():
  print(r)
```

```python
countries = ['Finland', 'South Korea']
df_plot = df_2022[df_2022['Country'].isin(countries)]
columns = ['Dystopia (1.83) + residual', 'Explained by: GDP per capita',
           'Explained by: Social support', 'Explained by: Healthy life expectancy',
           'Explained by: Freedom to make life choices',
           'Explained by: Generosity', 'Explained by: Perceptions of corruption']
num_vars = len(columns)
angles = np.linspace(0, 2 * np.pi, num_vars, endpoint=False).tolist()
angles += angles[:1]
fig, ax = plt.subplots(figsize=(6, 6), subplot_kw=dict(polar=True))
for k, (idx, row) in enumerate(df_plot.iterrows()):
    data = row[columns].tolist() + row[columns].tolist()[:1]
    ax.fill(angles, data, alpha=0.4, label=row['Country'], color=colors_mix[3*k], edgecolor=colors_mix[3*k])
    ax.plot(angles, data)
ax.set_xticks(angles[:-1])
col_name = ['Dystopia', 'GDP per capita',
           'Social support', 'Healthy life expectancy',
           'Freedom to make life choices',
           'Generosity', 'Perceptions of corruption']
ax.set_xticklabels(col_name)
plt.title('Comparison of Happiness Factors: Finland vs. South Korea', size=15, color=colors_dark[1], y=1.1)
ax.legend(loc='upper right', bbox_to_anchor=(1.1, 1.1))
plt.show()
```

```python
health_df = df_2022.sort_values(by='Explained by: Healthy life expectancy', ascending=False).reset_index().drop(columns=['index'])
health_df = health_df.reset_index().rename(columns={'index':'Healthy Rank'})
health_df['Healthy Rank'] = health_df['Healthy Rank'] + 1
```

```python
df = health_df[health_df['Country'].isin(Negara_Maju)]
fig, ax = plt.subplots(figsize=(14, 8))
bars1 = ax.bar(df['Country'], df['Explained by: Healthy life expectancy'], color=colors_dark[3], alpha=0.4, edgecolor=colors_dark[0])
ax.set_xticklabels(df['Country'], rotation=45, ha="right")
ax.set_xlabel("Countries", fontsize=14, labelpad=10, fontweight='bold', color=colors_dark[0])
ax.set_ylabel("Healthy life score", fontsize=14, labelpad=10, fontweight='bold', color=colors_dark[0])
plt.show()
```

```python
df = health_df[health_df['Country'].isin(Negara_Maju)]
health_df_top = health_df.iloc[0:1]
median_score = health_df['Explained by: Healthy life expectancy'].median()
health_idx = list(health_df.index + 1)
fig, ax = plt.subplots(figsize=(14, 8))
bars1 = ax.bar(df['Country'], df['Explained by: Healthy life expectancy'], color=colors_dark[3], alpha=0.4, edgecolor=colors_dark[0])
bars1[0].set_alpha(1)
bars1[0].set_alpha(1)
bars1[0].set_color(colors_green[0])
bars1[0].set_edgecolor(colors_green[0])
bars1[18].set_alpha(1)
bars1[18].set_color(colors_green[3])
bars1[18].set_edgecolor(colors_green[3])
line  = ax.axhline(median_score, linestyle='--', color=colors_dark[2])
ax.legend(["Median"], loc='upper center', bbox_to_anchor=(0.5, -0.3), ncol=4, borderpad=1, frameon=False, fontsize=12)
ax.grid(axis='y', alpha=0.3)
ax.set_xticklabels(df['Country'], rotation=45, ha="right")
ax.set_axisbelow(True)
ax.set_xlabel("Countries", fontsize=14, labelpad=10, fontweight='bold', color=colors_dark[0])
ax.set_ylabel("Healthy life expectancy", fontsize=14, labelpad=10, fontweight='bold', color=colors_dark[0])
xmin, xmax = ax.get_xlim()
ymin, ymax = ax.get_ylim()
avgl = ax.text(
    s="Global\nMedian: {:.2f}".format(median_score),
    x=xmax*1.02,
    y=median_score,
    backgroundcolor=colors_dark[2],
    fontsize=12,
    fontweight='bold',
    color='white'
)
for i, bar in enumerate(bars1):
    x = bar.get_x()
    y = bar.get_height()
    if i not in [0,18]:
        ax.text(
            s=f"{health_idx[i]}th",
            va='center', ha='center',
            x=x+0.38, y=y/2,
            color=colors_dark[3],
            fontsize=12,
        )
    else:
        ax.text(
            s=f"{health_idx[i]}th",
            va='center', ha='center',
            x=x+0.38, y=y/2,
            color='white',
            fontsize=12,
            fontweight='bold'
        )
plt.text(s="Negara dengan harapan hidup tertinggi diantara negara maju ", ha='left', x=xmin, y=ymax*1.12, fontsize=24, fontweight='bold', color=colors_dark[0])
plt.title(f"Diantara negara maju, Jepang menempati posisi pertama dalam harapan hidup tertinggi.\nSedangkan Amerika menempati posisi terakhir.", loc='left', fontsize=13, color=colors_dark[2])
plt.tight_layout()
plt.show()
```
